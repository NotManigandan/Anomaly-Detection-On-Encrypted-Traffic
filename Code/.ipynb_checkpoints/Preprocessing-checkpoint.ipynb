```python
def expand_categories(values):
    result = []
    s = values.value_counts()
    t = float(len(values))
    for v in s.index:
        result.append("{}:{}%".format(v,round(100*(s[v]/t),2)))
    return "[{}]".format(",".join(result))
        
def analyze(df):
    cols = df.columns.values
    total = float(len(df))
    for col in cols:
        uniques = df[col].unique()
        unique_count = len(uniques)
        if unique_count>100:
            print("--> {}:{} ({}%)".format(col,unique_count,int(((unique_count)/total)*100)))
        else:
            print("--> {}:{}".format(col,expand_categories(df[col])))
            expand_categories(df[col])
```

```python
analyze(df)
```

# Feature Selection

```python
X = df.drop(columns=['Label'], axis = 1)
y = df[['Label']]
```

```python
# Dropping features with zero variance before they have same valuea for all samples which won't help the model in anyway
statistics = X.describe()
std_dev = statistics.loc['std'] # The describe() gives statisctical info in which std gives standard deviation info
features_zero_variance = std_dev[std_dev == 0.0].index # Finding column with 0 Variance
print("Features with 0 Variance:-")
print(pd.Series(features_zero_variance).sort_values())
X = X.drop(columns=features_zero_variance)
print(f"Dataframe shape after removing features with zero variance: {X.shape}")
```

```python
# Correlation Analysis
correlation_matrix = spearmanr(X).correlation
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(25, 40))
sns.heatmap(correlation_matrix, xticklabels=X.columns, yticklabels=X.columns, linewidths=.5, cmap=sns.diverging_palette(620, 10, as_cmap=True), ax=ax1)

# Hierarchical clustering on the correlation matrix using ward method which is an aggloremative (average distance) clustering method
# Ward is much suitable for quantitative variables
correlation_linkage = hierarchy.ward(correlation_matrix)
dendro = hierarchy.dendrogram(correlation_linkage, labels=X.columns, ax=ax2)
cluster_threshold = 1 # We need to remove highly correlated features
# We select 1 because the lower the threshold or the lower the distance, the more similar they are and distance of 1 will give us higly correlated eatures. Any number more thhan this will remove necessary features
ax2.plot([0, 12345], [cluster_threshold, cluster_threshold], ':r')
plt.show()
```

```python
cluster_ids = hierarchy.fcluster(correlation_linkage, cluster_threshold, criterion='distance')
cluster_id_to_feature_ids = defaultdict(list)

# For each cluster under the threshold, we keep only the first feature and remove the other corrrelated feature
for idx, cluster_id in enumerate(cluster_ids):
    cluster_id_to_feature_ids[cluster_id].append(idx)
selected_features = [v[0] for v in cluster_id_to_feature_ids.values()]

selected_features = X.columns[selected_features].tolist()

print('Selected features:')
print(pd.Series(selected_features))

X = X[selected_features]
print(f"Number of Selected Features features: {len(X.columns)}")
```

```python
corr_updated = spearmanr(X).correlation
plt.figure(figsize=(25, 20))
sns.heatmap(corr_updated, xticklabels=X.columns, yticklabels=X.columns, linewidths=.5, cmap=sns.diverging_palette(620, 10, as_cmap=True))
```

```python
import matplotlib.pyplot as plt

# Model names
labels = ['Model 1', 'Model 2', 'Model 3', 'Model 4', 'Model 5']

# Position of bars
x = np.arange(len(labels))  

# Width of the bars
width = 0.35  

accuracy_unencrypted = [0.9, 0.85, 0.88, 0.92, 0.95]
accuracy_encrypted = [0.85, 0.82, 0.86, 0.88, 0.9]

recall_unencrypted = [0.9, 0.85, 0.88, 0.92, 0.94]
recall_encrypted = [0.85, 0.82, 0.86, 0.88, 0.9]

f1_score_unencrypted = [0.93, 0.91, 0.9, 0.92, 0.96]
f1_score_encrypted = [0.88, 0.9, 0.88, 0.87, 0.9]

precision_unencrypted = [0.91, 0.93, 0.94, 0.95, 0.97]
precision_encrypted = [0.9, 0.89, 0.92, 0.91, 0.93]

fig, axs = plt.subplots(2, 2, figsize=(15, 10))  # create a subplot with 2 rows and 2 columns

# Flatten the axs array to easily iterate over it
axs = axs.flatten()

# List of metrics for easier iteration
metrics_unencrypted = [accuracy_unencrypted, f1_score_unencrypted, recall_unencrypted, precision_unencrypted]
metrics_encrypted = [accuracy_encrypted, f1_score_encrypted, recall_encrypted, precision_encrypted]
metric_names = ['Accuracy', 'F1 Score', 'Recall', 'Precision']

# Create bar plots
for i in range(4):
    axs[i].bar(x - width/2, metrics_unencrypted[i], width, label='Unencrypted')
    axs[i].bar(x + width/2, metrics_encrypted[i], width, label='Encrypted')
    axs[i].set_ylabel(metric_names[i])
    axs[i].set_xticks(x)
    axs[i].set_xticklabels(labels)
    axs[i].legend()

# Add a title and display the plot
fig.suptitle('Model performance on Unencrypted vs Encrypted data')
plt.show()
```
